# AR1モデル
$y(t)$ $= $$\mu'$ + $\phi y(t-1)$ + $e(t)$,  $e(t) \sim \mathcal{N}(0, \sigma^2)$
を実装し、シミュレーションを行う
なお、AR(1)モデルではe(t)はホワイトノイズであることを満たせば正規分布である必要はない

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#シミュレーション最大期間の設定
simulation_time = 1000
#シミュレーション結果を格納する変数の作成
vy = np.zeros(simulation_time)
```

```python
#mu,sigmaを設定する
mu = float(input())
sigma = float(input())
```

```python
#定常性の判定と初期値の設定
phi = float(input())
if -1 < phi < 1:
  print("phiは定常性をみたしています")
  vy[0]=mu/(1-phi)
else:
  print("phiは定常性を満たしていません")
  vy[0]=0
```

```python
#実際にAR(1)モデルを指定された期間適用する
for t in range(simulation_time-1):
  vy[t+1] = mu + phi * vy[t] + np.random.normal(0,sigma)
```

```python
#シミュレーション結果を描写する
plt.plot(vy,color='salmon')
plt.hlines(vy[0], 0, simulation_time, colors='steelblue', linestyle='dashed')
plt.xlabel('Time')
plt.ylabel('y(t)')
```
